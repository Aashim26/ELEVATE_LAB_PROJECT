DATA CLEANING

```python

# CUSTOMER CHURN ANALYSIS
# STEP 2 - DATA CLEANING
# ==========================================

# Import Libraries
import pandas as pd
import numpy as np
import os

# ==========================================
# LOAD DATASET
# ==========================================

df = pd.read_csv("C:\\Users\\91890\\Desktop\\Customer_Churn_Analysis\\dataset\\WA_Fn-UseC_-Telco-Customer-Churn.csv")

print("="*50)
print("DATASET LOADED SUCCESSFULLY")
print("="*50)

# ==========================================
# DATASET SHAPE
# ==========================================

print("\nDataset Shape:")
print(df.shape)

# ==========================================
# FIRST 5 ROWS
# ==========================================

print("\nFirst 5 Rows:")
print(df.head())

# ==========================================
# DATASET INFO
# ==========================================

print("\nDataset Information:")
print(df.info())

# ==========================================
# CHECK MISSING VALUES
# ==========================================

print("\nMissing Values:")
print(df.isnull().sum())

# ==========================================
# CHECK BLANK VALUES IN TotalCharges
# ==========================================

print("\nBlank Values in TotalCharges:")
print((df["TotalCharges"] == " ").sum())

# ==========================================
# CONVERT TotalCharges TO NUMERIC
# ==========================================

df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)

# ==========================================
# CHECK MISSING VALUES AFTER CONVERSION
# ==========================================

print("\nMissing Values After Conversion:")
print(df.isnull().sum())

# ==========================================
# FILL MISSING VALUES WITH MEDIAN
# ==========================================

median_value = df["TotalCharges"].median()

df["TotalCharges"] = df["TotalCharges"].fillna(
    median_value
)

print("\nMissing Values After Filling:")
print(df.isnull().sum())

# ==========================================
# CHECK DUPLICATES
# ==========================================

print("\nDuplicate Records:")
print(df.duplicated().sum())

# ==========================================
# REMOVE DUPLICATES IF ANY
# ==========================================

df = df.drop_duplicates()

# ==========================================
# CHECK DATA TYPES
# ==========================================

print("\nData Types:")
print(df.dtypes)

# ==========================================
# SUMMARY STATISTICS
# ==========================================

print("\nSummary Statistics:")
print(df.describe())

# ==========================================
# SAVE CLEANED DATASET
# ==========================================

df.to_csv(
    "cleaned_telco_churn.csv",
    index=False
)

print("\nCleaned Dataset Saved Successfully!")

# ==========================================
# FINAL SHAPE
# ==========================================

print("\nFinal Dataset Shape:")
print(df.shape)

print("\nDATA CLEANING COMPLETED SUCCESSFULLY")
```

FEATURE ENGINEERING

```python
# ==========================================
# CUSTOMER CHURN ANALYSIS
# STEP 4 - FEATURE ENGINEERING
# ==========================================

# Import Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# ==========================================
# LOAD CLEANED DATASET
# ==========================================

df = pd.read_csv("cleaned_telco_churn.csv")

print("Dataset Shape:", df.shape)

# ==========================================
# DROP CUSTOMER ID
# ==========================================

df.drop("customerID", axis=1, inplace=True)

print("\ncustomerID removed successfully")

# ==========================================
# LABEL ENCODING
# ==========================================

le = LabelEncoder()

for column in df.columns:
    if df[column].dtype == "object":
        df[column] = le.fit_transform(df[column])

print("\nCategorical columns encoded successfully")

# ==========================================
# CHECK DATA TYPES
# ==========================================

print("\nData Types:")
print(df.dtypes)

# ==========================================
# TARGET VARIABLE
# ==========================================

X = df.drop("Churn", axis=1)
y = df["Churn"]

print("\nFeature Shape:", X.shape)
print("Target Shape:", y.shape)

# ==========================================
# TRAIN TEST SPLIT
# ==========================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("\nTrain Shape:", X_train.shape)
print("Test Shape:", X_test.shape)

# ==========================================
# FEATURE IMPORTANCE CHECK
# ==========================================

correlation = df.corr()["Churn"].sort_values(ascending=False)

print("\nCorrelation with Churn:")
print(correlation)

# ==========================================
# SAVE FEATURE ENGINEERED DATA
# ==========================================

feature_df = pd.concat([X, y], axis=1)

feature_df.to_csv(
    "feature_engineered_telco_churn.csv",
    index=False
)

print("\nFeature Engineered Dataset Saved Successfully")

# ==========================================
# SAVE TRAIN TEST DATA
# ==========================================

X_train.to_csv("X_train.csv", index=False)
X_test.to_csv("X_test.csv", index=False)

y_train.to_csv("y_train.csv", index=False)
y_test.to_csv("y_test.csv", index=False)

print("\nTrain/Test Files Saved Successfully")

# ==========================================
# FEATURE SUMMARY
# ==========================================

print("\n===== FEATURE ENGINEERING SUMMARY =====")
print("1. customerID removed")
print("2. Categorical columns encoded")
print("3. Features and target separated")
print("4. Train-Test Split completed")
print("5. Dataset ready for Machine Learning")

print("\nFEATURE ENGINEERING COMPLETED SUCCESSFULLY")
```

MODEL BUILDING

```python
# ==========================================
# CUSTOMER CHURN ANALYSIS
# STEP 5 - MODEL BUILDING
# ==========================================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

# ==========================================
# LOAD TRAIN TEST DATA
# ==========================================

X_train = pd.read_csv("X_train.csv")
X_test = pd.read_csv("X_test.csv")

y_train = pd.read_csv("y_train.csv").squeeze()
y_test = pd.read_csv("y_test.csv").squeeze()

print("Data Loaded Successfully")

# ==========================================
# FUNCTION TO EVALUATE MODEL
# ==========================================

def evaluate_model(model, model_name):

    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    print("\n" + "="*50)
    print(model_name)
    print("="*50)

    print("Accuracy :", round(accuracy,4))
    print("Precision:", round(precision,4))
    print("Recall   :", round(recall,4))
    print("F1 Score :", round(f1,4))

    print("\nClassification Report")
    print(classification_report(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(5,4))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues"
    )

    plt.title(f"{model_name} Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    return accuracy


# ==========================================
# LOGISTIC REGRESSION
# ==========================================

lr_model = LogisticRegression(
    max_iter=1000,
    random_state=42
)

lr_model.fit(X_train, y_train)

lr_accuracy = evaluate_model(
    lr_model,
    "Logistic Regression"
)

# ==========================================
# DECISION TREE
# ==========================================

dt_model = DecisionTreeClassifier(
    random_state=42
)

dt_model.fit(X_train, y_train)

dt_accuracy = evaluate_model(
    dt_model,
    "Decision Tree"
)

# ==========================================
# RANDOM FOREST
# ==========================================

rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train, y_train)

rf_accuracy = evaluate_model(
    rf_model,
    "Random Forest"
)

# ==========================================
# MODEL COMPARISON
# ==========================================

results = pd.DataFrame({
    "Model":[
        "Logistic Regression",
        "Decision Tree",
        "Random Forest"
    ],
    "Accuracy":[
        lr_accuracy,
        dt_accuracy,
        rf_accuracy
    ]
})

print("\n")
print("="*50)
print("MODEL COMPARISON")
print("="*50)
print(results)

# ==========================================
# ACCURACY BAR CHART
# ==========================================

plt.figure(figsize=(8,5))

sns.barplot(
    x="Model",
    y="Accuracy",
    data=results
)

plt.title("Model Accuracy Comparison")
plt.ylim(0.70,1.00)

plt.show()

# ==========================================
# BEST MODEL
# ==========================================

best_model = results.loc[
    results["Accuracy"].idxmax()
]

print("\nBest Model:")
print(best_model)

# ==========================================
# SAVE BEST MODEL
# ==========================================

import joblib

joblib.dump(
    rf_model,
    "random_forest_model.pkl"
)

print("\nModel Saved Successfully!")

print("\nMODEL BUILDING COMPLETED SUCCESSFULLY")
```

```python
# ==========================================
# CUSTOMER CHURN ANALYSIS
# STEP 6 - FEATURE IMPORTANCE ANALYSIS
# ==========================================

# Import Libraries
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================
# LOAD TRAINING DATA
# ==========================================

X_train = pd.read_csv("X_train.csv")

print("X_train Loaded Successfully")
print("Shape :", X_train.shape)

# ==========================================
# LOAD TRAINED MODEL
# ==========================================

model = joblib.load("random_forest_model.pkl")

print("Random Forest Model Loaded Successfully")

# ==========================================
# EXTRACT FEATURE IMPORTANCE
# ==========================================

feature_importance = pd.DataFrame({
    "Feature": X_train.columns,
    "Importance": model.feature_importances_
})

# Sort Descending
feature_importance = feature_importance.sort_values(
    by="Importance",
    ascending=False
)

print("\nTop 10 Important Features")
print(feature_importance.head(10))

# ==========================================
# SAVE FEATURE IMPORTANCE CSV
# ==========================================

feature_importance.to_csv(
    "feature_importance.csv",
    index=False
)

print("\nfeature_importance.csv Saved Successfully")

# ==========================================
# TOP 10 FEATURE IMPORTANCE CHART
# ==========================================

plt.figure(figsize=(10,6))

sns.barplot(
    data=feature_importance.head(10),
    x="Importance",
    y="Feature"
)

plt.title("Top 10 Features Influencing Customer Churn")
plt.xlabel("Importance Score")
plt.ylabel("Features")

plt.tight_layout()
plt.show()

# ==========================================
# FULL FEATURE IMPORTANCE CHART
# ==========================================

plt.figure(figsize=(12,8))

sns.barplot(
    data=feature_importance,
    x="Importance",
    y="Feature"
)

plt.title("Feature Importance Ranking")
plt.xlabel("Importance Score")
plt.ylabel("Features")

plt.tight_layout()
plt.show()

# ==========================================
# TOP CHURN FACTORS
# ==========================================

print("\n")
print("="*50)
print("TOP CHURN FACTORS")
print("="*50)

for i in range(min(10, len(feature_importance))):
    feature = feature_importance.iloc[i]["Feature"]
    score = feature_importance.iloc[i]["Importance"]

    print(f"{i+1}. {feature} --> {score:.4f}")

# ==========================================
# BUSINESS RECOMMENDATIONS
# ==========================================

print("\n")
print("="*50)
print("BUSINESS RECOMMENDATIONS")
print("="*50)

print("1. Encourage customers to choose long-term contracts.")
print("2. Provide retention offers for high-risk customers.")
print("3. Improve technical support services.")
print("4. Offer discounts for customers with high monthly charges.")
print("5. Create loyalty programs for long-tenure customers.")
print("6. Monitor customers likely to churn and engage them proactively.")

print("\nAnalysis Completed Successfully!")
```
